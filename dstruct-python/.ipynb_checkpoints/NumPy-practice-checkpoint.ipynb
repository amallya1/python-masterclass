```python
import matplotlib.pyplot as plt;
import numpy as np;
from math import *;
%matplotlib inline





```

```python
x = np.linspace(0,5,100)
print (x)
y = np.sin(4*x)*np.exp(-x)
z = np.cos(4*x)*np.exp(-x)
plt.figure(1)
plt.plot(x,y,'r--',label='$e^{-x} sin(4x)$')
plt.plot(x,z,label='$e^(-x)cos(4x)$')
plt.ylabel('value')
plt.xlabel('x')
plt.legend()

```
